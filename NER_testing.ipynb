```python
max_sent_tokens = 50
test_prop = 0.1
dev_prop = 0.1
```

```python
import os
import re
import us
import sys
import json

from nltk import sent_tokenize
from nltk import word_tokenize
from random import shuffle

data_path = '/media/rbshaffer/My Paasport/Current_Materials/dissertation/Legislative_Data/Legislation/UnitedStates/Annual'

file_list = os.listdir(data_path)

sys.path.append('/home/rbshaffer/PycharmProjects/Legislative_Data/')
from _country_entities_annual import UnitedStates

with open('/home/rbshaffer/Desktop/agency_list.txt', 'r') as f:
    agency_list = f.read().split('\n')
    
agency_list = [e for e in agency_list if e and e not in [s.name for s in us.states.STATES]]
```

```python
def process_inds(inst_list, sentence_str, tokens_list, label, tag_set, label_list):
    for inst_str in inst_list:
        if inst_str + ' ' in sentence_str or ' ' + inst_str in sentence_str:
            entity_len = len(inst_str.split())
            match_inds = [i for i in range(len(tokens_list)) 
                          if tokens_list[i:i + entity_len] == inst_str.split()]
            for i in match_inds:
                if i not in tag_set:
                    tag_set[i] = 'B-MISC'
                    
                    label_list.append(label)
                if entity_len > 1:
                    for j in range(i+1, i+entity_len):
                        if j not in tag_set:
                            tag_set[j] = 'I-MISC'
    
    return tag_set, label_list
```

```python
train_set = []
dev_set = []
test_set = []
full_set = []

agency_inds = list(range(len(agency_list)))
shuffle(agency_inds)

dev_start = int((1-(dev_prop + test_prop))*len(agency_list))
test_start = int((1-(test_prop))*len(agency_list))

train_strs = [agency_list[i] for i in agency_inds[:dev_start]]
dev_strs = [agency_list[i] for i in agency_inds[dev_start:test_start]]
test_strs = [agency_list[i] for i in agency_inds[test_start:]]

for file_name in file_list:
    print(re.sub('_', '/', file_name).strip('.json'))

    with open(data_path + '/' + file_name, 'r') as f:
        content = json.loads(f.read())

        if content['parsed']:
            # imported from system path above
            parser = UnitedStates(content['parsed'])
            for chunk in parser.chunks:
                sentences = sent_tokenize(chunk)
                for sentence in sentences:
                    # SUPER IMPORTANT STEP
                    sentence = re.sub('\s+', ' ', sentence)
                    tags = {}
                    labels = []
                    
                    tokens = word_tokenize(sentence)
                    
                    tags, labels = process_inds(train_strs, sentence, tokens, 'train', tags, labels)
                    tags, labels = process_inds(dev_strs, sentence, tokens, 'dev', tags, labels)
                    tags, labels = process_inds(test_strs, sentence, tokens, 'test', tags, labels)                        
                    
                    tags.update({i: 'O' for i in range(len(tokens)) if i not in tags.keys()})
                    to_append = '\n'.join(tokens[i] + ' ' + tags[i] for i in range(len(tokens)))
                    
                    if 'train' in labels and 'dev' not in labels and 'test' not in labels:
                        train_set.append(to_append)
                    elif 'dev' in labels and 'train' not in labels and 'test' not in labels:
                        dev_set.append(to_append)
                    elif 'test' in labels and 'dev' not in labels and 'train' not in labels:
                        test_set.append(to_append)
                    else:
                        full_set.append(to_append)      
                    
```

```python
train_examples = [e for e in train_set if len(e.split()) < max_sent_tokens and 'MISC' in e]
shuffle(train_examples)

dev_examples = [e for e in dev_set if len(e.split()) < max_sent_tokens and 'MISC' in e]
shuffle(dev_examples)

test_examples = [e for e in test_set if 'MISC' in e]
shuffle(test_examples)

with open('/home/rbshaffer/PycharmProjects/Legislative_Data/data_v1_train.txt', 'w') as f:
    f.write('\n\n'.join(train_examples))
    
with open('/home/rbshaffer/PycharmProjects/Legislative_Data/data_v1_dev.txt', 'w') as f:
    f.write('\n\n'.join(dev_examples))
    
with open('/home/rbshaffer/PycharmProjects/Legislative_Data/data_v1_test.txt', 'w') as f:
    f.write('\n\n'.join(test_examples))
```
